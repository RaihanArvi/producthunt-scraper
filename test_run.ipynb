# Test Run for Scraping One Day

This code tests the scraper by scraping a single date, or a single product.

```python
from producthunt_scraper.core.script import *
```

```python
# Scrape the daily leaderboard for a given date
process_date = datetime(2026, 2, 1)

browser = await nd.start()
products = await scrape_products(browser, process_date)
```

```python
# Pick the first product
prod = products[3]
print(prod.model_dump(mode='json'))
```

```python
# Test scraping the first product only.
browser = await nd.start()
product = await scrape_single_product(browser, prod)
print(product.model_dump(mode='json'))
```

```python
# Scrape all products in the given day.
browser = await nd.start()

for prod in products:
    product = await scrape_single_product(browser, prod)
    print(product.model_dump(mode='json'))
```
